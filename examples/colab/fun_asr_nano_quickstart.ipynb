## 2. Basic Recognition

Transcribe audio in Chinese, English, or Japanese with a single API call.

```python
from funasr import AutoModel

model = AutoModel(
    model="FunAudioLLM/Fun-ASR-Nano-2512",
    trust_remote_code=True,
    device="cuda:0",
    hub="hf"
)

# Use built-in example audio
wav_path = f"{model.model_path}/example/zh.mp3"
result = model.generate(input=[wav_path], batch_size=1)
print(result[0]["text"])
```

## 3. With Timestamps

```python
result = model.generate(input=[wav_path], batch_size=1)
# Timestamps are included in the output
for item in result:
    print(f"Text: {item['text']}")
    if 'timestamp' in item:
        print(f"Timestamps: {item['timestamp']}")
```

## 4. With Hotwords

Boost recognition of domain-specific terms.

```python
result = model.generate(
    input=[wav_path],
    batch_size=1,
    hotwords=["开放时间", "博物馆"]
)
print(result[0]["text"])
```

## 5. With Speaker Diarization

Add VAD + speaker models for who-said-what.

```python
model_spk = AutoModel(
    model="FunAudioLLM/Fun-ASR-Nano-2512",
    trust_remote_code=True,
    device="cuda:0",
    hub="hf",
    vad_model="fsmn-vad",
    spk_model="cam++",
    punc_model="ct-punc"
)

result = model_spk.generate(input=[wav_path], batch_size=1)
for item in result:
    if 'sentence_info' in item:
        for sent in item['sentence_info']:
            spk = sent.get('spk', '?')
            text = sent.get('text', '')
            start = sent.get('start', 0)
            end = sent.get('end', 0)
            print(f"[{start}ms-{end}ms] Speaker {spk}: {text}")
```

## 6. Try Your Own Audio

Upload a file or record from microphone.

```python
from google.colab import files
uploaded = files.upload()

for filename in uploaded.keys():
    result = model.generate(input=[filename], batch_size=1)
    print(f"\n--- {filename} ---")
    print(result[0]["text"])
```
